```python
import os
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data_path = 'D:/SegEM/volumeTrainingData/'
file_name = '1.mat'

file_path = os.path.join(data_path, file_name)

mat_file = sio.loadmat(file_path)
mat_file.keys()
```

```python
target = mat_file['target']
raw = mat_file['raw']
```

```python
print(mat_file['__header__'])
print(mat_file['__version__'])
print(mat_file['__globals__'])
```

```python
plt.subplot(1, 2, 1)
plt.imshow(raw[:, :, 0], cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(target[:, :, 0], cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show
```

```python
# Plotting several images from the array
plt.figure(figsize=(20, 10))

for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(raw[:, :, i], cmap='gray')
    plt.axis('off')
    plt.title('Raw{}'.format(i), fontsize=20)

for i in range(5):
    plt.subplot(2, 5, i+6)
    plt.imshow(target[:, :, i], cmap='gray')
    plt.axis('off')
    plt.title('Target{}'.format(i), fontsize=20)
    
plt.tight_layout()
plt.show()
```
