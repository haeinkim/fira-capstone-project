```python
import os
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

# 1. /volumeTrainingData

## Load single image saved as .mat

```python
data_path = 'D:/SegEM/volumeTrainingData/'
file_name = '1005.mat'

file_path = os.path.join(data_path, file_name)

mat_file = sio.loadmat(file_path)
mat_file.keys()
```

```python
target = mat_file['target']
raw = mat_file['raw']
```

### Raw and target label has different shape of array

```python
print(raw.shape,target.shape)
```

## Pixel coordinates quite different between raw and label images

```python
plt.subplot(1, 2, 1)
plt.imshow(raw[:,:,0], cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(target[:,:,0], cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show
```

## Table S2 gives pixel coordinates information for raw and label images

```python
bbox = np.array([[2200,2200,2000],
                 [2300,2300,2100]])

x_interval = bbox[1][0] - bbox[0][0]
y_interval = bbox[1][1] - bbox[0][1]
z_interval = bbox[1][2] - bbox[0][2]

print(x_interval, y_interval, z_interval)
```

```python
bboxRaw = np.array([[2150,2150,1975], 
                    [2350,2350,2125]])

x_interval = bboxRaw[1][0] - bboxRaw[0][0]
y_interval = bboxRaw[1][1] - bboxRaw[0][1]
z_interval = bboxRaw[1][2] - bboxRaw[0][2]

print(x_interval, y_interval, z_interval)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(raw[50:150, 50:150, 25], cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(target[:,:,0], cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(20, 10))

for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.imshow(raw[50:150, 50:150, 25 + i], cmap='gray')
    plt.axis('off')
    plt.title('Raw{}'.format(i), fontsize=20)

for i in range(5):
    plt.subplot(2, 5, i+6)
    plt.imshow(target[:, :, i], cmap='gray')
    plt.axis('off')
    plt.title('Target{}'.format(i), fontsize=20)
    
plt.tight_layout()
plt.show()
```
